```python
import numpy as np
import matplotlib.pyplot as plt

# Generación de datos sintéticos
np.random.seed(0) # Para reproducibilidad
X = np.random.rand(100, 1) * 10  # 100 puntos de datos en el rango de 0 a 10
coef_real = 1.5
intercepto_real = 2
ruido = np.random.randn(100, 1) * 2  # Ruido gaussiano
y_real = intercepto_real + coef_real * X + ruido

# Simulación de predicciones de un modelo
y_pred = intercepto_real + coef_real * X + np.random.randn(100, 1) * 0.5  # Predicciones con menos ruido
```

## Visualización de los datos y de las predicciones

```python
# Visualización
plt.figure(figsize=(8, 6))
plt.scatter(X, y_real, color='pink', label='Datos reales')
plt.scatter(X, y_pred, color='lightblue', label='Predicciones')
plt.plot(X, intercepto_real + coef_real * X, color='black', label='Línea de regresión')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Visualización de datos y predicciones')
plt.legend()
plt.show()
```

## Cálculo del MSE y R²

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_real, y_pred)
r2 = r2_score(y_real, y_pred)

print(f"The Mean Squared Error (MSE) is: {mse:.2f}")
print(f"The R-squared (R2) score is: {r2:.2f}")
```
