```python
import numpy as np
import matplotlib.pyplot as plt

# Generación de datos sintéticos
np.random.seed(0) # Para reproducibilidad
X = np.random.rand(100, 1) * 10  # 100 puntos de datos en el rango de 0 a 10
coef_real = 1.5
intercepto_real = 2
ruido = np.random.randn(100, 1) * 2  # Ruido gaussiano
y_real = intercepto_real + coef_real * X + ruido

# Simulación de predicciones de un modelo
y_pred = intercepto_real + coef_real * X + np.random.randn(100, 1) * 0.5  # Predicciones con menos ruido
```

## Visualización de los datos y de las predicciones

```python
# Visualización de los valores reales vs. predicciones
plt.figure(figsize=(10, 6))
plt.scatter(X, y_real, color='blue', label='Valores Reales')
plt.scatter(X, y_pred, color='red', alpha=0.5, label='Predicciones')
plt.title('Comparación de Valores Reales y Predicciones')
plt.xlabel('Variable Independiente X')
plt.ylabel('Variable Dependiente Y')
plt.legend()
plt.show()
```

## Cálculo del MSE y R²

```python
from sklearn.metrics import mean_squared_error, r2_score

# Cálculo del MSE y R²
mse = mean_squared_error(y_real, y_pred)
r2 = r2_score(y_real, y_pred)

print(f"MSE: {mse}\nR2: {r2}")
```
