```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

X, _ = make_blobs(n_samples=300, centers=4, random_state=42, cluster_std=1.0)
```

Ayuda: [Documentación KMeans](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html)

## Visualización de los clusters

```python
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

## Cálculo del valor de inertia con scikit-learn (max 10 clusters)

```python
# Número máximo de clusters a probar
max_clusters = 10
```

```python
# Usar el atributo inertia de scikit-learn
inertia_sklearn = []
for k in range(1, max_clusters + 1):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
    kmeans.fit(X)
    inertia_sklearn.append(kmeans.inertia_)

plt.plot(range(1, max_clusters + 1), inertia_sklearn, marker='o', color='orange')
plt.title('Método del Codo (Atributo inertia_)')
plt.xlabel('Número de Clusters (k)')
plt.ylabel('Inertia')

plt.show()
```
