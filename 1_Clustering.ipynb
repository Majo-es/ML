```python
# Importamos las libreras necesarias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generamos datos sintticos para clustering
X_cluster, y_cluster = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Visualizacin del clustering
plt.figure(figsize=(10, 6))
plt.scatter(X_cluster[:, 0], X_cluster[:, 1])
plt.title('Aprendizaje No Supervisado de Clustering')
plt.xlabel('Caracterstica 1')
plt.ylabel('Caracterstica 2')
plt.show()
```

```python
# Creamos y entrenamos el modelo KMeans
model_cluster = KMeans()
model_cluster.fit(X_cluster)

# Predecimos los clusters
y_pred_cluster = model_cluster.predict(X_cluster)

# Visualizacin del clustering
plt.figure(figsize=(10, 6))
plt.scatter(X_cluster[:, 0], X_cluster[:, 1], c=y_pred_cluster, cmap='viridis')
plt.title('Aprendizaje No Supervisado de Clustering')
plt.xlabel('Caracterstica 1')
plt.ylabel('Caracterstica 2')
plt.show()
```
