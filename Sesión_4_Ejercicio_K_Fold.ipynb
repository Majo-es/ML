````python
k_fold = KFold(n_splits=5)

model = LinearRegression()

scores = cross_val_score(model, X, y, cv=k_fold, scoring="neg_mean_squared_error") # cv: Cross-validation object

print("Results of cross validation (MSE):")
for i, score in enumerate (scores, start=1):
  print(f"fold {i}: {score:.4f}")

# CALCULATE
mean_accuracy = scores.mean()
print(f"\nPrecision promedio: {mean_accuracy:.4f}")
````

Using  R-squared (R^2) score

````python
import numpy as np

# 5 FOLD CROSS-VALIDATION
kf = KFold(n_splits=5, shuffle=True, random_state=42)

scores = []
for train_index, test_index in kf.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    # ADJUST THE MODEL
    model = LinearRegression()
    model.fit(X_train, y_train)

    score = model.score(X_test, y_test)
    scores.append(score)

# R2
avg_score = np.mean(scores)
print(f"Puntaje R-cuadrado promedio: {avg_score:.4f}")
````
