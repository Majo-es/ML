```python
# Importamos las librerías necesarias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Generamos datos sintéticos para una regresión lineal simple
X_reg = np.random.rand(100, 1) * 10  # Característica aleatoria
y_reg = 2 * X_reg + np.random.randn(100, 1) * 2  # Respuesta con ruido

# Visualización de la regresión
plt.figure(figsize=(10, 6))
plt.scatter(X_reg, y_reg, color='blue', label='Datos')
plt.title('Aprendizaje Supervisado de Regresión')
plt.xlabel('Característica')
plt.ylabel('Respuesta')
plt.legend()
plt.show()
```

```python
# Dividimos los datos en entrenamiento y prueba
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)

# Creamos y entrenamos el modelo de regresión lineal
model_reg = LinearRegression()
model_reg.fit(X_train_reg, y_train_reg) #The fit() method adjusts the model parameters based on the input data (X) and the target values (y).

# Realizamos predicciones y evaluamos el modelo
y_pred_reg = model_reg.predict(X_test_reg)

# Visualización de la regresión
plt.figure(figsize=(10, 6))
plt.scatter(X_reg, y_reg, color='blue', label='Datos')
plt.plot(X_test_reg, y_pred_reg, color='red', label='Línea de regresión')
plt.title('Aprendizaje Supervisado de Regresión')
plt.xlabel('Característica')
plt.ylabel('Respuesta')
plt.legend()
plt.show()
```

```python
new_data = input("Introduce un número:")
test = [float(new_data)]
test_predict = model_reg.predict([test])

plt.figure(figsize=(10, 6))
plt.scatter(X_reg, y_reg, color='blue', label='Datos')
plt.scatter(test, test_predict, color='green', label='Dato nuevo')
plt.plot(X_test_reg, y_pred_reg, color='red', label='Línea de regresión')
plt.title('Aprendizaje Supervisado de Regresión')
plt.xlabel('Característica')
plt.ylabel('Respuesta')
plt.legend()
plt.show()
```
