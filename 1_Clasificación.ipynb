```python
# Importamos las libreras necesarias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_blobs
from sklearn.tree import DecisionTreeClassifier

# Generamos datos sintticos para clustering
X_cluster, y_cluster = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Visualizacin del clustering
plt.figure(figsize=(10, 6))
plt.scatter(X_cluster[:, 0], X_cluster[:, 1], c=y_cluster, s=50, cmap='viridis')
plt.title('Aprendizaje Supervisado de Clasificacin')
plt.xlabel('Caracterstica 1')
plt.ylabel('Caracterstica 2')
plt.show()
```

```python
# Dividimos los datos en entrenamiento y prueba
X_train_clas, X_test_clas, y_train_clas, y_test_clas = train_test_split(X_cluster, y_cluster, test_size=0.2, random_state=42)

# Visualizacin de la clasificacin
plt.figure(figsize=(10, 6))
plt.scatter(X_test_clas[:, 0], X_test_clas[:, 1], cmap='viridis')
plt.title('Aprendizaje Supervisado de Clasificacin')
plt.xlabel('Caracterstica 1')
plt.ylabel('Caracterstica 2')
plt.show()
```

```python
# Creamos y entrenamos el modelo de clasificacin
model_clas = DecisionTreeClassifier()
model_clas.fit(X_train_clas, y_train_clas)

# Evaluamos el modelo
y_pred_clas = model_clas.predict(X_test_clas)

# Visualizacin de la clasificacin
plt.figure(figsize=(10, 6))
plt.scatter(X_test_clas[:, 0], X_test_clas[:, 1], c=y_pred_clas, cmap='viridis')
plt.title('Aprendizaje Supervisado de Clasificacin')
plt.xlabel('Caracterstica 1')
plt.ylabel('Caracterstica 2')
plt.show()
```
