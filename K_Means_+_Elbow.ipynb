## Carga de datos

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

X, y = make_blobs(n_samples=300, centers=3, random_state=42, cluster_std=1.0)
```

Ayuda: [Documentación KMeans](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html)

## Visualización de los datos

```python
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.title('Original Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

## Aplicar K-Means seleccionando el valor óptimo de K

```python
max_clusters = 10
# Usar el atributo inertia de scikit-learn
inertia_sklearn = []
for k in range(1, max_clusters + 1):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertia_sklearn.append(kmeans.inertia_)

plt.plot(range(1, max_clusters + 1), inertia_sklearn, marker='o', color='orange')
plt.title('Método del Codo (Atributo inertia_)')
plt.xlabel('Número de Clusters (k)')
plt.ylabel('Inertia')

plt.show()
```

## Visualización de los centroides con el valor óptimo de K

```python
kmeans = KMeans(n_clusters=3, random_state=42)
y_kmeans = kmeans.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis', edgecolors='k', s=50)
plt.title('Dataset After K-Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis', edgecolors='k', s=50)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='hotpink', marker='X', label='Centroids')
plt.title('Dataset After K-Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```
