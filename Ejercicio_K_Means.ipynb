## Carga de datos

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# When X is uppercase is a Matrix
# When x is lowercase is a vector

X, y  = make_blobs(n_samples=300, centers=4, random_state=42, cluster_std=1.0) # random_state defines the starting point
```

Ayuda: [Documentación KMeans](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html)

## Visualización de los datos

```python
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.title('Original Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

## Visualización de los clusters originales

```python
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', s=50)
plt.title('Original Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

## Visualización de los clusters después de aplicar K-Means

```python
# Fit the K-Means model
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)


# Visualize the data and the clusters
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis', s=50)
plt.title('Dataset After K-Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

## Visualización de los centroides

```python
# Visualize the data and the clusters
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis', s=50)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='hotpink',  marker='X', label='Centroids')
plt.title('Dataset After K-Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
print('Cluster Centroids:')
for i, center in enumerate(kmeans.cluster_centers_):
    print(f'Cluster {i}: ({center[0]:.2f}, {center[1]:.2f})')
```
