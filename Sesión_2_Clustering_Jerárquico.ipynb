```python
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
```

## Clustering Jerárquico Aglomerativo

```python
# Generar un conjunto de datos aleatorio
X, _ = make_blobs(n_samples=150, centers=4, n_features=2, random_state=40)

# Visualización
plt.scatter(X[:, 0], X[:, 1])
plt.title('Clustering Jerárquico Aglomerativo')
plt.xlabel('Característica 1')
plt.ylabel('Característica 2')
plt.show()
```

```python
# Aplicar clustering jerárquico aglomerativo
clustering = AgglomerativeClustering(n_clusters=4)
clustering.fit(X)

# Visualización
plt.scatter(X[:, 0], X[:, 1], c=clustering.labels_, cmap='rainbow')
plt.title('Clustering Jerárquico Aglomerativo')
plt.xlabel('Característica 1')
plt.ylabel('Característica 2')
plt.show()
```

## Clusterign Jerárquico Divisivo

```python
import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from matplotlib import pyplot as plt
from sklearn.datasets import make_blobs

# Generar un conjunto de datos
X, _ = make_blobs(n_samples=150, centers=4, n_features=2, random_state=40)

# Visualización
plt.scatter(X[:, 0], X[:, 1])
plt.title('Clustering Jerárquico Divisivo')
plt.xlabel('Característica 1')
plt.ylabel('Característica 2')
plt.show()
```

```python
# Aplicar clustering jerárquico divisivo
linked = linkage(X, method='ward')

# Crear clusters (corte del dendrograma)
max_d = 7.5  # define el criterio de corte
clusters = fcluster(linked, max_d, criterion='distance')

# Visualización
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap='rainbow')
plt.title('Clustering Jerárquico Divisivo')
plt.xlabel('Característica 1')
plt.ylabel('Característica 2')
plt.show()
```

### [Manhattan & Euclidean distance ](https://https://medium.com/analytics-vidhya/euclidean-and-manhattan-distance-metrics-in-machine-learning-a5942a8c9f2f)
> Euclidean Distance represents the shortest distance between two points.

> Manhattan Distance is the sum of absolute differences between points across all the dimensions.



image.png

CALCULATING THE MANHATTAN DISTANCE

```python
# Calculation using NumPy arrays:
import numpy as np
point_a_np = np.array([1, 1, 1])
point_b_np = np.array([4, 5, 6])
# np.sum()->  representing the Manhattan distance between the two points.
distance_numpy = np.sum(np.abs(point_a_np - point_b_np)) #NP.ABS -> returns the absolute values of each element
print(f"Manhattan distance (NumPy): {distance_numpy}")
```

```python
# Calculation using SciPy’s cityblock() function:
from scipy.spatial.distance import cityblock
point_a = (1, 1, 1)
point_b = (4, 5, 6)
distance_scipy = cityblock(point_a, point_b)
print(f"Manhattan distance (SciPy): {distance_scipy}")
```

Calculating the Euclidean Distance

```python
import numpy as np
from scipy.spatial.distance import euclidean
def euclidean_distance(point1, point2):
    return np.sqrt(np.sum((np.array(point1) - np.array(point2))**2))

# 2D example
point_a = (1, 2)
point_b = (4, 6)
distance_2d = euclidean_distance(point_a, point_b)
print(f"2D Euclidean distance: {distance_2d:.2f}")

# 3D example
point_c = (1, 2, 3)
point_d = (4, 6, 8)
distance_3d = euclidean_distance(point_c, point_d)
print(f"3D Euclidean distance: {distance_3d:.2f}")

# Using SciPy for efficiency
distance_scipy = euclidean(point_c, point_d)
print(f"3D Euclidean distance (SciPy): {distance_scipy:.2f}")
```
